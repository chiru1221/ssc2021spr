```python
NAME = 'lgb_group'
feature_importance = dict()
feature_name = list()
cv_num = 10
for i in range(cv_num):
    for j in range(cv_num):
        model = lgb.Booster(model_file='../model/{0}_{1}_{2}.txt'.format(NAME, i, j))
        if i == 0 and j == 0:
            feature_importance = {name: importance / 100.0 for name, importance in zip(model.feature_name(), model.feature_importance())}
            feature_name = model.feature_name()
        else:
            for k, name in enumerate(feature_name):
                feature_importance[name] += model.feature_importance()[k] / 100.0
```
